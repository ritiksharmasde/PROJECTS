```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress


x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])


slope, intercept, r_value, p_value, std_err = linregress(x, y)


y_fit = slope * x + intercept

plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_fit, color='red', label=f'Fit Line: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x = np.array([1, 2, 3, 4, 5])
y = np.array([1.1, 2.0, 4.5, 8.0, 15.5])


def exponential(x, a, b):
    return a * np.exp(b * x)

params, _ = curve_fit(exponential, x, y)
a, b = params

y_fit = exponential(x, a, b)


plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_fit, color='red', label=f'Fit Curve: y = {a:.2f}e^({b:.2f}x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
